```python
import os
from random import choice
import shutil
import kagglehub
```

```python
path = kagglehub.dataset_download("fxmikf/diverse-lpd-training-ready")

print("Path to dataset files:", path)
```

```python
dataset_dir = "/kaggle/input/diverse-lpd-training-ready"

if os.path.exists(dataset_dir):
    drive_dest_dir = "/content/Gdrive/MyDrive/Data-input-colab"
    os.makedirs(drive_dest_dir, exist_ok=True)

    for item in os.listdir(dataset_dir):
        src_path = os.path.join(dataset_dir, item)
        dst_path = os.path.join(drive_dest_dir, item)

        if os.path.isdir(src_path):
            shutil.copytree(src_path, dst_path)
        else:
            shutil.copy(src_path, dst_path)

    print("Contents copied to Google Drive")
else:
    print("Dataset directory does not exist.")
```

```python
imgs =[]
xmls =[]

train_path = "/content/yolov5/data/images/train"
val_path = "/content/yolov5/data/images/val"
source_image = "/content/Gdrive/MyDrive/Data-input-colab/Diverse-LPDv2/images"
source_label = "/content/Gdrive/MyDrive/Data-input-colab/Diverse-LPDv2/labels"

if not os.path.exists(train_path):
  os.mkdir(train_path)
if not os.path.exists(val_path):
  os.mkdir(val_path)

train_ratio = 0.8
val_ratio = 0.2

totalImgCount = len(os.listdir(source_image))

for (dirname, dirs, files) in os.walk(source_image):
  for filename in files:
    imgs.append(filename)

for (dirname, dirs, files) in os.walk(source_label):
  for filename in files:
    xmls.append(filename)

countForTrain = int(len(imgs)*train_ratio)
countForVal = int(len(imgs)*val_ratio)
print("training images are:", countForTrain)
print("validation images are:", countForVal)
```
